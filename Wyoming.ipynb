```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
```

```python
def actionss(xpaths):
    but = driver.find_element(By.XPATH,xpaths)
    action.move_to_element(but).perform()
    action.click(but).perform()
```

```python
driver = webdriver.Chrome()
action = ActionChains(driver)
driver.maximize_window()
driver.get('https://findchildcarewy.org/maps/')
time.sleep(5)
```

```python
facility_search = driver.find_element(By.XPATH,'//*[@id="divSplash"]/div[2]/div/div[1]')
action.move_to_element(facility_search).perform()
action.click(facility_search).perform()
time.sleep(1)
```

```python
text_version = driver.find_element(By.XPATH,'//*[@id="divFacilitySearch"]/div[1]')
action.move_to_element(text_version).perform()
action.click(text_version).perform()
time.sleep(1)
```

```python
child_care_center = driver.find_element(By.XPATH,'//*[@id="divSearchChildCareCenter"]')
action.move_to_element(child_care_center).perform()
action.click(child_care_center).perform()
```

```python
List_result = driver.find_element(By.XPATH,'//*[@id="divSearchParameters"]/div[14]/div/div')
action.move_to_element(List_result).perform()
action.click(List_result).perform()
time.sleep(3)
```

```python
License_number = []
Provider_Name = []
City = []
Zip_Code = []
Address = []
Phone_Number = []
Email = []
Capacity = []
```

```python
actionss('//*[@id="divListResultsTarget"]/table/tbody/tr/td[4]/div')
```

```python
x=1
while x<15:
    
    lic = driver.find_elements(By.XPATH,'//*[@id="tblResults"]/tbody/tr/td[1]/div')
    for l in lic:
        action.move_to_element(l).perform()
        License_number.append(l.text)   
    data = driver.find_elements(By.XPATH,'//*[@id="tblResults"]/tbody/tr/td[2]/div')
    for d in data:
        action.move_to_element(d).perform()
        action.click(d).perform()
        time.sleep(5)
        n = driver.find_element(By.XPATH,'//*[@id="facilitydetails"]/span')
        Provider_Name.append(n.text)
        a = driver.find_element(By.XPATH,'//*[@id="facilitydetails"]')
        textcon = a.text
        textcon = textcon.split('\n')
        Address.append(textcon[1])
        print(Address)
        zip_data = textcon[2].split(", WY ")
        try:
            Zip_Code.append(zip_data[1])
            City.append(zip_data[0])
            Phone_Number.append(textcon[3])
        except:
            Zip_Code.append("not found")
            City.append("not found")
            Phone_Number.append("not found")
        try:
            e = driver.find_element(By.XPATH,'//*[@id="facilitydetails"]/table[1]/tbody/tr/td[1]/a')
            Email.append(e.get_attribute('href'))
        except:
            Email.append("not found")
        c = driver.find_element(By.XPATH,'//*[@id="divPopupFacilityOverview"]')
        con = c.text
        con = con.split("\n")
        Capacity.append(con[1])
        action.send_keys(Keys.ESCAPE).perform()
        time.sleep(2)  
    
    actionss('//*[@id="divListResultsTarget"]/table/tbody/tr/td[4]/div')
    time.sleep(5)
    x = x+1
    
    
```
