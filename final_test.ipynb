```python
center_name = []
overall_rating = []
no_of_people_rated = []
center_type = []
address = []
website = []
contact = []
plus_code = []
operating_hours = []
rewiewer_list = [[] for _ in range(5)]
rewiewer_rating = [[] for _ in range(5)]
rewiewer_review = [[] for _ in range(5)]
rewiewer_time = [[] for _ in range(5)]
```

```python
# Set up the options for the Chrome WebDriver
chrome_options = webdriver.ChromeOptions()
# You can add additional options if needed, such as specifying the Chrome binary location.

# Create the Chrome WebDriver instance using the specified options
driver = webdriver.Chrome(options=chrome_options)

# Now you can use the driver to navigate to a website
driver.get("https://www.google.com/maps/@31.117853,74.440827,15z?entry=ttu")
time.sleep(5)
```

```python
for i in range(len(list_colorado)):
    import time
    location = driver.find_element(By.XPATH, '//*[@id="searchboxinput"]')
    location.click()
 # Clear input field
    location.clear()
    time.sleep(5)
    location.send_keys(list_colorado[i])
    time.sleep(5)


    search_button = driver.find_element(By.XPATH, '//*[@id="searchbox-searchbutton"]')
    search_button.click()
    time.sleep(10)
    
    try:
        n = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div/div[1]/div[1]/h1')
        time.sleep(1)
        center_name.append(n.text)
    except:
        center_name.append("not found")
    time.sleep(3)    
        
        
    try:
        r = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div/div[1]/div[2]/div/div[1]/div[2]/span[1]/span[1]')
        time.sleep(1)
        overall_rating.append(r.text)
    except:
        overall_rating.append("not found")
    time.sleep(3)  
        
        
    try:
        no = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div/div[1]/div[2]/div/div[1]/div[2]/span[2]/span/span')
        time.sleep(1)
        no_of_people_rated.append(no.text.replace('(', '').replace(')', ''))
    except:
        no_of_people_rated.append("not found")
    time.sleep(3)
    
    try:
        ct = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div/div[1]/div[2]/div/div[2]/span/span/button')
        time.sleep(1)
        center_type.append(ct.text)
    except:
        center_type.append("not found")
    time.sleep(3)
    
    
    try:
        a = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[7]/div[3]/button/div/div[2]/div[1]')
        time.sleep(1)
        address.append(a.text)
    except:
        address.append("not found")
    time.sleep(3)
    
    try:
        w = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[7]/div[6]/a/div/div[2]/div[1]')
        time.sleep(1)
        website.append(w.text)
    except:
        website.append("not found")
    time.sleep(3)
    
    try:
        con = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[7]/div[7]/button/div/div[2]/div[1]')
        time.sleep(1)
        contact.append(con.text)
    except:
        contact.append("not found")
    time.sleep(3)
    
    try:
        pc = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[7]/div[8]/button/div/div[2]/div[1]')
        time.sleep(1)
        plus_code.append(pc.text)
    except:
        plus_code.append("not found")
    time.sleep(3)
    
    try:
        table_element = driver.find_element(By.CLASS_NAME, 'mxowUb')
        aria_label_value =table_element.get_attribute('aria-label')
        time.sleep(1)
        operating_hours.append(aria_label_value)
    except:
        operating_hours.append("not found")
    time.sleep(8)
    
    
    try:
        review_button = driver.find_element(By.XPATH,'/html/body/div[3]/div[8]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[3]/div/div/button[2]')
        review_button.click()
        time.sleep(3)
        sort_button = driver.find_element(By.XPATH,'//button[@aria-label="Sort reviews"]')
        sort_button.click()
        time.sleep(3)
        newest_click = driver.find_element(By.XPATH,'/html/body/div[3]/div[3]/div[1]/div[2]')
        newest_click.click()
        time.sleep(2)
    except:
        pass
    
    
    try:
        WebDriverWait(driver, 15).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class="jftiEf fontBodyMedium "]')))
    except:
        pass
    
    try:
        more_button = driver.find_elements(By.XPATH,'//button[@class="w8nwRe kyuRq"]')
        for button in more_button:
            button.click()
        WebDriverWait(driver, 15).until(EC.element_to_be_clickable((By.XPATH,'//button[@class="w8nwRe kyuRq"]')))
    except:
        pass
    # driver.implicitly_wait(10)
    try:
        WebDriverWait(driver, 15).until(EC.find_elements((By.XPATH,'//div[@class="jftiEf fontBodyMedium "]')))
    except:
        pass

    time.sleep(10)
    soup = BeautifulSoup(driver.page_source, 'lxml')

    names = soup.find_all('div', attrs={'class': 'd4r55'})
    review_time = soup.find_all('span', attrs={'class': 'rsqaWe'})
    try:
        WebDriverWait(driver, 15).until(EC.find_elements((By.XPATH,'//div[@class="jftiEf fontBodyMedium "]')))
    except:
        pass
    review = soup.find_all('span', attrs={'class': 'wiI7pd'})
    span_elements = soup.find_all('span', class_='kvMYJc')

    rewiewer_name = []
    for name in names:
        try:
            rewiewer_name.append(name.text)
        except:
            rewiewer_name.append('Not Found')

        if len(rewiewer_name) >= 5:
            break
    #     else:
    #         rewiewer_name.append('Not Found')


    time_review = []       
    for time in review_time:
        try:
            time_review.append(time.text)
        except:
            time_review.append('Not Found')

        # Break the loop after appending the first 5 ratings
        if len(time_review) >= 5:
            break
    #     else:
    #         time_review.append('Not Found')


    rating = []
    for span_element in span_elements:
        try:
            aria_label_value = span_element['aria-label']
            rating.append(aria_label_value)
        except:
            rating.append('Not Found')

            # Break the loop after appending the first 5 ratings
        if len(rating) >= 5:
            break
    #     else:
    #         rating.append('Not Found')


    reviews = []
    for item in review:
        try:
            reviews.append(item.text)
        except:
            reviews.append('Not Found')

        # Break the loop after appending the first 5 ratings
        if len(reviews) >= 5:
            break
    #     else:
    #         reviews.append('Not Found')


    if not rewiewer_name:  # Check if reviewer_name is empty
        for j in range(5):
            rewiewer_list[j].append('Not Found')        
    else:        
        for i, name in enumerate(rewiewer_name):
            if name:
                rewiewer_list[i % 5].append(name)

        for j in range(len(rewiewer_name), 5):
            rewiewer_list[j].append(np.nan)


    if not rating:  # Check if reviewer_name is empty
        for j in range(5):
            rewiewer_rating[j].append('Not Found')        
    else:
        for i, rate in enumerate(rating):
            if rate:
                rewiewer_rating[i%5].append(rate)


        for j in range(len(rating), 5):
            rewiewer_rating[j].append(np.nan)


    if not reviews:  # Check if reviewer_name is empty
        for j in range(5):
            rewiewer_review[j].append('Not Found')        
    else:        
        for i, review in enumerate(reviews):
            rewiewer_review[i%5].append(review)


        for j in range(len(reviews), 5):
            rewiewer_review[j].append(np.nan)


    if not time_review:  # Check if reviewer_name is empty
        for j in range(5):
            rewiewer_time[j].append('Not Found')        
    else:
        for i, time in enumerate(time_review):
            rewiewer_time[i%5].append(time)   

        for j in range(len(time_review), 5):
            rewiewer_time[j].append(np.nan)
```

```python
import re

def clean_special_characters(input_list):
    cleaned_list = []

    for item in input_list:
        cleaned_item = re.sub(r'[^\w\s]', '', re.sub('\u202f', '', item))
        cleaned_list.append(cleaned_item)

    return cleaned_list
```
