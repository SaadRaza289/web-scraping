```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
```

```python
driver = webdriver.Chrome()
action = ActionChains(driver)
driver.maximize_window()
```

```python
def waitelem(links):
    timeout = 10
    a=1
    element_found = False
    while not element_found:
        try:
            # Wait until the element with id "myElement" appears
            element = WebDriverWait(driver, timeout).until(
                EC.presence_of_element_located((By.XPATH,links)))

            # If the element appears, set the flag to True and exit the loop
            element_found = True

            # Perform further actions with the element
            print("Element appeared!")
            # Additional code to interact with the element
        except:
            # Handle the exception when the element does not appear within the specified timeout
            print("Element not found. Retrying...")
            a=a+1
        if a ==3:
            element_found = True
```

```python
def selection(xpaths, name):
    dropdown_element = driver.find_element(By.XPATH,xpaths)

    # Create a Select object with the dropdown element
    dropdown = Select(dropdown_element)

    dropdown.select_by_visible_text(name)
```

```python
def actionss(xpaths):
    but = driver.find_element(By.XPATH,xpaths)
    action.move_to_element(but).perform()
    action.click(but).perform()
```

```python
driver.get('https://childcarefind.okdhs.org/providers?provider-name=a&facility-type=childcare-center')
```

```python
links = driver.find_elements(By.XPATH,'//*[@id="radix-:R2e6:"]/ul/li/div/a')
```

```python
all_links = []
```

```python
for l in links:
    all_links.append(l.get_attribute('href'))
```

```python
len(all_links)
```

```python
Name = []
License_Number = []
Level = []
Capacity = []
Phone_Number = []
Email = []
Address = []
City = []
Zip_Code = []
```

```python
for li in all_links:
    driver.get(li)
    waitelem('//*[@id="__next"]/main/section/h1')
    try:
        n = driver.find_element(By.XPATH,'//*[@id="__next"]/main/section/h1')
        Name.append(n.text)
    except:
        Name.append("not found")
    
    try:
        l = driver.find_element(By.XPATH,'//*[@id="__next"]/main/section/div')
        l = l.text
        l = l.split('- ')
        License_Number.append(l[1])
    except:
        License_Number.append('not found')
        
    try:
        le = driver.find_element(By.XPATH,'//*[@id="__next"]/main/section/h2')
        Level.append(le.text)
    except:
        Level.append("not found")
    
    try:
        c = driver.find_element(By.XPATH,'//*[@id="__next"]/main/div/div[1]/section[2]/section[2]')
        c = c.text
        c = c.split('\n')
        Capacity.append(c[1])
    except:
        Capacity.append("not found")
    
    try:
        p = driver.find_element(By.XPATH,'//*[@id="__next"]/main/div/div[2]/div/div/div[1]')
        Phone_Number.append(p.text)
    except:
        Phone_Number.append("not found")
    
    try:
        e = driver.find_element(By.XPATH,'//*[@id="__next"]/main/div/div[2]/div/div/div[2]')
        Email.append(e.text)
    except:
        Email.append("not found")
        
    try:
        data = driver.find_element(By.XPATH,'//*[@id="__next"]/main/div/div[2]/div/div/div[3]')
        data = data.text
        data1 = data.split('\n')
        Address.append(data1[0])
        data2 = data1[1]
        data2 = data2.split(', OK ')
        City.append(data2[0])
        Zip_Code.append(data2[1])
    except:
        Address.append("not found") 
        City.append("not found")
        Zip_Code.append("not found")
```
